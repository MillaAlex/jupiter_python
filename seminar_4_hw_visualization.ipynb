## Условие 3: 3 задача
### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

1. Взаимосвязь стоимости дома и года постройки

```python
plt.figure(figsize=(25,7))
sns.boxplot(x=df['yr_built'], y=df['price'], showfliers=False)
plt.xlabel('Год постройки')
plt.ylabel('Стоимость дома')
plt.xticks(rotation=97)
plt.title('Взаимосвязь стоимости дома и года постройки');
```

Вывод: стоимость домов примерно в одном ценовом диапазоне в зависимости от года постройки, за исключением домов, построенных в 1933 году - здесь стоимость домов значительно выше.

2. Взаимосвязь стоимости дома и состояния.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['condition'], y=df['price'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и состояния')
plt.xlabel('Состояние')
plt.ylabel('Стоимость дома')
plt.grid();
```

Вывод: стоимость дома напрямую зависит от его состояния.

3. Взаимосвязь стоимости дома и количества спален.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['bedrooms'], y=df['price'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость дома')
plt.grid();
```

Вывод: Стоимость дома в некоторой степени зависит от количества спален. Стоимость домов повышается, где количество спален до 8, далее стоимость не зависит от этого показателя.

4. Взаимосвязь стоимости дома и хорошего вида.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['view'], y=df['price'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и хорошего вида')
plt.xlabel('Показатель вида')
plt.ylabel('Стоимость дома')
plt.grid();
```

Вывод: стоимость дома сильно зависит от показателя вида: чем выше показатель вида, тем выше стоимость дома.

5. Взаимосвязь стоимости дома и ремонта дома.

```python
plt.figure(figsize=(25,6))
sns.boxplot(x=df['yr_renovated'], y=df['price'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и ремонта дома')
plt.xlabel('Год ремонта дома')
plt.ylabel('Стоимость дома')
plt.xticks(rotation=97)
plt.grid();
```

Вывод: стоимость дома скорее не зависит от ремонта, а зависит от года постройки дома.

6. Взаимосвязь стоимости дома и количества этажей.

```python
plt.figure(figsize=(25,6))
sns.boxplot(x=df['floors'], y=df['price'], showfliers=False)
plt.title('Взаимосвязь стоимости дома и количества этажей')
plt.xlabel('К-во этажей')
plt.ylabel('Стоимость дома')
plt.grid();
```

Вывод: стоимость дома с 2.5 этажей выше стоимости остальных домов.

7. Тепловая матрица корреляций.

```python
corr_m = df.corr().round(1)
corr_m[np.abs(corr_m) < 0.3] = 0
corr_m
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr_m, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Матрица корреляций')
```
