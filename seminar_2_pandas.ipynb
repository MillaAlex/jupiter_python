#### Как много ноутбуков Ultrabook с 8GB RAM?


```python
ultrabook = df[df['TypeName'] == 'Ultrabook'].shape[0]
ultrabook
```

```python
ultra8 = df[(df['TypeName'] == 'Ultrabook') & (df['Ram'] == '8GB')].shape[0]
ultra8
```

```python
res = round(ultra8 / ultrabook * 100, 2)
res
```

```python
(df[df['TypeName'] == 'Ultrabook'].value_counts('Ram', normalize=True).loc['8GB'] * 100).round(2)
```

#### Выберите ноутбук клиенту. Клиент хочет подобрать ноутбук с 8GB или 16GB ОЗУ на Windows 10 в стоимости до 500 евро, сколько у него вариантов?

#### Выберите ноутбук клиенту. Клиент хочет подобрать ноутбук от MSI, с видеокартой Nvidia GeForce GTX 1050 Ti и главное не с диагональю 15.6. В какой ценовой категории вышли подобные ноутбуки?

```python
chosen = df[(df['Company'] == 'MSI') & (df['Gpu'] == 'Nvidia GeForce GTX 1050 Ti') & (df['Inches'] != 15.6)]['Price_euros']
chosen
```

```python
chosen.min(), chosen.max()
```
